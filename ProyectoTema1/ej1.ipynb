```python
import numpy as np
import scipy as scp
import matplotlib.pyplot as plt
import random as rnd

REAL = 1 - scp.stats.norm.cdf(3) # 1 - P(X <= 3)

f = lambda y : (2*scp.pi)**(-1/2) * np.exp(-(((1/y) + 2)**2)/2) * y**(-2)

def ejercicio1a(n):
    ''' Metodo de montecarlo en [0,1]
    '''
    acc = 0
    for _ in range(n):
        acc += f(rnd.random())
    return acc/n
```

```python
equid = np.arange(10_000,500_001,10_000)
acc = {}
for i in equid:
    acc[i] = round(ejercicio1a(i),7-1) 
    print(f"n: {i:7} val: {acc[i]}")
```

```python
plt.figure(figsize=(20,8))
plt.xticks(equid, rotation=45)
plt.plot([1000,500_000],[REAL,REAL], label=)
plt.plot(acc.keys(), acc.values())
plt.show()
```
