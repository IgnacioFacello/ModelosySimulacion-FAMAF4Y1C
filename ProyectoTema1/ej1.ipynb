```python
import numpy as np
from scipy.stats import norm, uniform
import matplotlib.pyplot as plt
import random as rnd

EQUID = np.arange(10_000,500_001,10_000)
REAL = 1 - norm.cdf(3) # 1 - P(X <= 3)
```

```python
def plot_aproximation(acc)
    plt.figure(figsize=(20,8))
    plt.xticks(EQUID, rotation=45)
    plt.plot([1000,500_000],[REAL,REAL], label='real')
    plt.plot(acc.keys(), acc.values())
```

## Inciso 1

$$
f(y) = (2*\pi)**(-1/2) * e^{-({(1/y) + 2}^2)\over 2} * y^(-2)
$$

```python
f = lambda y : (2*np.pi)**(-1/2) * np.exp(-(((1/y) + 2)**2)/2) * y**(-2)

def ejercicio1a(n):
    ''' Metodo de montecarlo en [0,1]
    '''
    acc = 0
    ys = uniform.scv(size=n)
    for y in ys:
        acc += f(y)
    return acc/n
```

```python
acc_mtc = {}
for i in EQUID:
    acc_mtc[i] = round(ejercicio1a(i),6)
    # print(f'{i:7} : {acc[i]}')
```

```python
plot_aproximation(acc_mtc)
plt.show()
```
