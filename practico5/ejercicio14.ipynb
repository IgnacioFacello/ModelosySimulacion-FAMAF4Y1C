```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def capacity():
    return 20 + int(rnd.random() * 21)

def proc_poisson(tt, lam):
    t = 0
    nt = 0
    eventos = []
    while t < tt:
        t += - np.log(1 - rnd.random()) / lam
        if t <= tt:
            nt += 1
            eventos.append((t,capacity()))
    return nt, eventos
```

```python
def autobuses(hs):
    return proc_poisson(hs, 5)
```

```python
plt.figure(figsize=(18,8))

rnd.seed(9090)

n = 1 # horas a simular

t, axes = autobuses(n)

xs = [ x[0]*60 for x in axes]
vals = [ x[1] for x in axes]
x_full = np.arange(0,60*n,1)
y_acc = []
for i in x_full:
    temp =  [ b for a, b in axes if a*60 < i ]
    y_acc.append(sum(temp))

plt.bar(xs,vals, color='red')
plt.plot(x_full,y_acc, color='blue')

plt.xticks(x_full)
plt.show()
```

En rojo las llegadas y en azul el total de hinchas que llegaron
