## Ejercicio a
$$ 
F(x) =
\begin{cases}
0 \\
\int^{x}_{2} {t \over 2} - 1 dt & 2 \le x \lt 3 \\
\int^{3}_{2} {t \over 2} - 1 dt + \int^{x}_{3} 1 - {t \over 6} dt & 3 \le x \lt 6\\
1
\end{cases}
$$

---

$$ \begin{align}
\int^{x}_{2} {t-2 \over 2} dt \\
{t^2\over 4} - x \bigg |^{x}_{2} \\
{x^2\over 4} - x + 1 \\
{x^2 - 4x + 4 \over 4}  
\end{align} $$

---

$$ \begin{align}
\int^{x}_{3} 1 - {t \over 6} dt \\
t - {t^2\over 12} \bigg |^{x}_{3} \\
 \\
\end{align} $$

---

$$ 
F(x) =
\begin{cases}
0 \\
{x^2 - 4x + 4 \over 4} & 2 \le x \lt 3 \\
-{x^2 - 12x + 24\over 12} & 3 \le x \lt 6\\
1
\end{cases}
$$

```python
import matplotlib.pyplot as plt
import numpy as np

def f_mass(x):
    if 2 <= x < 3:
        return (x-2)/2
    elif 3 <= x < 6:
        return (2-x/3)/2
    else:
        return 0

def f_acc(x):
    if x < 2:
        return 0
    elif 2 <= x < 3:
        return (x**2/4) - x + 1
    elif 3 <= x < 6:
        return - (x**2/12) + x - 2
    else:
        return 1

def map_func(f, ls):
    return list(map(f, ls))

step = 0.5
x_vals = np.arange(1, 8, step)
plt.plot(x_vals, map_func(f_acc, x_vals), label="Acumulada")
plt.plot(x_vals, map_func(f_mass, x_vals), label="Masa")
print(np.sum(map_func(f_mass, x_vals))*step)
plt.grid(linestyle = '--', linewidth = 1)
plt.legend()
plt.show()
```

## Ejercicio b

$$
F(x) =
\begin{cases}
0 & x \lt 0\\
\int^{x}_{0} {6t + 18 \over 35} dt & 0 \le x \lt 1 \\
\int^{1}_{0} {6t + 18 \over 35} dt + \int^{x}_{1} {6t^2 \over 35 } dt & 1 \le x \lt 2 \\
1 & x \ge 2 \\
\end{cases}
$$
---
$$\begin{align}
\int^{x}_{0} {6t + 18 \over 35} dt \\   
\left. {3t^2 + 18t\over35} \right\rvert_{0}^{x} \\
3t^2 + 18t\over35
\end{align}$$
---
$$\begin{align}
\int^{x}_{1} {6t^2 \over 35 } dt \\   
\left. {3t^2\over35} \right\rvert_{1}^{x} \\
{3t^2 - 3\over35}
\end{align}$$
---
$$
F(x) =
\begin{cases}
0 & x \lt 0\\
{3x^2 + 18x\over 35} & 0 \le x \lt 1 \\
 {3x^2 + 18\over35} & 1 \le x \lt 2 \\
1 & x \ge 2 \\
\end{cases}
$$

```python
import matplotlib.pyplot as plt
import numpy as np

def f_mass(x):
    if 0 <= x < 1:
        return (6*x + 18*x) / 37
    elif 1 <= x < 2:
        return 6*x**2 / 35
    else:
        return 0

def f_acc(x):
    if x < 0:
        return 0
    elif 0 <= x < 1:
        return (3*x**2 + 18*x)/35
    elif 1 <= x < 2:
        return (3*x**2 + 18)/35
    else:
        return 1

def map_func(f, ls):
    return list(map(f, ls))

step = 0.5
x_vals = np.arange(-1, 4, step)
plt.plot(x_vals, map_func(f_acc, x_vals), label="Acumulada")
plt.plot(x_vals, map_func(f_mass, x_vals), label="Masa")
print(np.sum(map_func(f_mass, x_vals))*step)
plt.grid(linestyle = '--', linewidth = 1)
plt.legend()
plt.show()
```
